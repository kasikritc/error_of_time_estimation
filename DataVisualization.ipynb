```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
matplotlib.use('nbagg')
```

```python
df = pd.read_csv('data_tidy.csv')
df
```

```python
itv1 = df[df['Interval'] == 1]
itv3 = df[df['Interval'] == 3]
itv5 = df[df['Interval'] == 5]
```

```python
fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
ax1.set(title='Error of Time Estimation',
       xlabel='Pressing Round', 
       ylabel='Time (s)')
plt.show()
```

```python
plt.scatter(itv1['Pressing Round'], itv1['Time'], color='lightblue', s=5)
plt.plot(range(1, 6), range(1, 6), color = 'blue', linewidth=2)

plt.scatter(itv3['Pressing Round'], itv3['Time'], color='lightgreen', s=5)
plt.plot(range(1, 6), range(3, 16, 3), color = 'green', linewidth=2)

plt.scatter(itv5['Pressing Round'], itv5['Time'], color= 'orange', s=5)
plt.plot(range(1, 6), range(5, 26, 5), color = 'red', linewidth=2)
```

```python
fig2 = plt.figure()
ax2 = fig2.add_subplot(111)
ax2.set(title='Error of Time Estimation',
       xlabel='Time (s)', 
       ylabel='Percentage Error')
plt.show()
```

```python
plt.scatter(itv1['Time'], itv1['PCT Error'], color='lightblue', s=10)
plt.scatter(itv3['Time'], itv3['PCT Error'], color='lightgreen', s=10)
plt.scatter(itv5['Time'], itv5['PCT Error'], color='orange', s=10)
```
