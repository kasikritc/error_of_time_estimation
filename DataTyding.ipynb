```python
import pandas as pd
import numpy as np
import random
```

```python
df = pd.read_csv('data.csv')
df.set_index('Person')
```

```python
df = df.melt(id_vars='Person',
        value_vars=[column for column in list(df.columns) if column != 'Person'],
        var_name='Type',
        value_name='Time')
df
```

```python
df = (df.sort_values(['Person', 'Type'])
        .reset_index())
df
```

```python
df = df.drop(['index'], axis=1)
df
```

```python
types = list(df['Type'])
rounds = []
for i, string in enumerate(types):
    rounds.append(string[3])
df['Pressing Round'] = rounds
df
```

```python
types = list(df['Type'])
intervals = []
for i, string in enumerate(types):
    intervals.append(str(string)[0])

df['Interval'] = intervals
df
```

```python
del df['Type']
```

```python
df.head()
```

```python
cols = df.columns.tolist()
cols = [cols[0]] + [cols[3]] + [cols[2]] + [cols[1]]
df = df[cols]
df.head()
```

```python
df.info()
```

```python
df['Interval'] = df['Interval'].astype(int)
df['Pressing Round'] = df['Pressing Round'].astype(int)
df.info()
```

```python
df['PCT Error'] = round(((df['Time'] - (df['Pressing Round']*df['Interval']))
                    /(df['Pressing Round']*df['Interval'])*100), 2)
```
